## 迭代计算评估指标

```python
accuracy = evaluate.load('accuracy')
for ref, pred in zip([0, 1, 1], [0, 1, 0]):
    accuracy.add(reference=ref, prediction=pred) # 使用add将一组数据加载进入accuracy对象
accuracy.compute()
```

```python
accuracy = evaluate.load('accuracy')
for refs, preds in zip([[0, 1], [1, 0]], [[0, 1], [1, 1]]):
    accuracy.add_batch(references=refs, predictions=preds) # 使用add_batch将多组数据加载进入accuracy对象
accuracy.compute()
```
